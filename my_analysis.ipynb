```python
import duckdb
from tabulate import tabulate

# Create a connection with DuckDB
con = duckdb.connect(database=':memory:')

loan_opportunities = \
  con.read_csv("https://raw.githubusercontent.com/BlackRock2308/response-test-technique-meilleurtaux/refs/heads/main/data_sources/data_samples/opportunity_test.csv")

result = con.execute("SELECT * FROM loan_opportunities").df()

print("******** Results ********")
print(tabulate(result.head(10), headers='keys', tablefmt='grid'))

con.close()
```

```python
import duckdb
from tabulate import tabulate

# Create a connection with DuckDB
con = duckdb.connect(database=':memory:')

loan_opportunities = \
  con.read_csv("https://raw.githubusercontent.com/BlackRock2308/response-test-technique-meilleurtaux/refs/heads/main/data_sources/data_samples/opportunity_test.csv")


result = con.execute("""
SELECT * EXCLUDE (Id, A_une_proposition_de_sa_banque__c, Age_emprunteur__c, Appel_Automatique__c, Avancement__c)
FROM loan_opportunities;
""").fetchdf()


print("******** Results ********")
print(tabulate(result.head(10), headers='keys', tablefmt='grid'))

con.close()
```
